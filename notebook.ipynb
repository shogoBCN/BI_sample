## 1.2. Definition of Metrics

### Customer Value Ratio (CVR)

To monitor the success of our sales strategies, we propose the **Customer Value Ratio (CVR)** as a key performance metric. The CVR is a measure of the average revenue generated per customer for each minute spent on a sales method. It's calculated by dividing the Average Revenue per Customer by the time spent per sales method, then multiplying the result by 1000.

In our calculations, we've made the following assumptions about time investment for each sales method, as outlined by our sales representative:

- Email: 5 minutes (2.5 minutes per email)
- Email + Call: 15 minutes (2.5 minutes setup, 2.5 minutes for email and 10 minutes for the call)
- Call: 30 minutes

These ratios reflect the efficiency of our sales methods in terms of revenue generation versus time investment. A high CVR suggests a sales method that delivers strong returns on time invested, while a low CVR could indicate an inefficient approach requiring review and potential adjustment.

Using the current data, we can establish **initial values** for the CVR as follows:

| Sales Method | CVR (Initial Value) |
| ------------ | ------------------ |
| Call         | **0.31** |
| Email        | **2.41** |
| Email + Call | **4.11** |

We recommend setting up a real-time dashboard using a business intelligence tool to continuously track the CVR. This will provide an instant snapshot of the efficacy of our sales methods, enabling us to make swift, data-driven decisions for sales strategy optimization.

The CVR, alongside the other insights gleaned from our analysis, will guide us towards maximizing our revenue and ensuring the most efficient use of our sales team's time.

```python
## cleaned up value inconsistencies 
df.replace({"email": "Email", "em + call": "Email + Call"}, inplace=True)
customers_per_method = df["sales_method"].value_counts(dropna=False)
display(customers_per_method)

hist_plot = sns.histplot(data=df, x="sales_method", bins=3, kde=False, color='#3498db')
hist_plot.set_title('Sales Method Distribution', fontsize=18, fontweight='bold')
hist_plot.set_xlabel('Sales Method')
hist_plot.set_ylabel('Number of Customers')

for p in hist_plot.patches: ## count annotations
    hist_plot.annotate(format(p.get_height(), '.0f'),
                       (p.get_x() + p.get_width() / 2., p.get_height()),
                       xytext=(0, 5),
                       textcoords='offset points')
plt.show()
```

```python
# Group by sales method and state, then sum revenue
# For each sales method, get the top 5 states by revenue
sales_by_state_and_method = df.groupby(['state', 'sales_method'])['revenue'].sum().reset_index()
top_states_by_method = sales_by_state_and_method.sort_values(['sales_method', 'revenue'], ascending=False).groupby('sales_method').head(5)
display(top_states_by_method)

plt.figure(figsize=(16, 8))
sns.countplot(data=df, x='state', order = df['state'].value_counts().index, palette='coolwarm')
plt.title('Customer Distribution by State')
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.xticks(rotation=90)
plt.show()

state_revenue = df.groupby('state')['revenue'].sum().reset_index()
plt.figure(figsize=(16, 8))
sns.barplot(data=state_revenue, x='state', y='revenue', order = state_revenue.sort_values('revenue',ascending = False).state, palette='coolwarm')
plt.title('Highest Revenue by State')
plt.xlabel('State')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Creating a copy of the dataframe and adding an "Overall" category to the "sales_method" column.
# Grouping the data by the sales method and getting descriptive statistics for the "revenue" column.
# This is to allow for the calculation of overall statistics in addition to per sales method.
df_summary_stats = df.copy()
df_summary_stats['sales_method'] = 'Overall'
df_summary_stats = pd.concat([df, df_summary_stats])
grouped_revenue_stats = df_summary_stats.groupby("sales_method")["revenue"].describe()
print("\nRevenue statistics by sales method:")
display(grouped_revenue_stats.T)

# Setting up a subplots grid for visualizations.
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Revenue Spread by Sales Method', fontsize=18, fontweight='bold')

# Plotting a box plot to show the distribution of revenue for each sales method.
box_plot = sns.boxplot(ax=axes[0], x="sales_method", y="revenue", data=df_summary_stats, palette="coolwarm")
axes[0].set_title('Box Plot', fontsize=14)
axes[0].set_xlabel('Sales Method')
axes[0].set_ylabel('Revenue')

# Plotting a violin plot for a more detailed view of the distribution of revenue for each sales method.
violin_plot = sns.violinplot(ax=axes[1], x="sales_method", y="revenue", data=df_summary_stats, palette="coolwarm", inner=None, alpha=0.7)
axes[1].set_title('Violin Plot', fontsize=14)
axes[1].set_xlabel('Sales Method')
axes[1].set_ylabel('Revenue')

# Plotting a histogram to visualize the frequency distribution of the revenue for each sales method.
df_hist = df_summary_stats[df_summary_stats['sales_method'] != 'Overall']
plt.figure(figsize=(8, 6))
plt.hist(df_hist['revenue'], bins=30, alpha=0.5, label='Overall')
for method in df_hist["sales_method"].unique():
    plt.hist(df_hist[df_hist['sales_method'] == method]['revenue'], bins=30, alpha=0.5, label=method)
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.title('Distribution of Revenue Overall and for Each Sales Method')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot a box plot for overall revenue
plt.figure(figsize=(5, 4))
sns.boxplot(y=df_summary_stats[df_summary_stats["sales_method"] == "Overall"]["revenue"], color="skyblue")
plt.title('Overall Revenue', fontsize=14)
plt.xlabel('Overall')
plt.ylabel('Revenue')
plt.show()
```

```python
# Multiply the count of each sales method by the corresponding time spent per method to obtain total time spent per method
time_by_sales_method = df["sales_method"].value_counts() * [5, 30, 15]
time_by_sales_method.index = ["Email", "Call", "Email + Call"]

# Count the number of customers per sales method
customers_per_method = df["sales_method"].value_counts(dropna=False)

# Calculate the total and average revenue per sales method
total_revenue_per_method = df.groupby("sales_method")["revenue"].sum()
average_revenue_per_method = total_revenue_per_method / customers_per_method

# Calculate the revenue per hour for each sales method
revenue_per_hour = total_revenue_per_method / time_by_sales_method * 60

# Calculate the Customer Value Ratio (CVR) for each sales method
CVR_per_method = average_revenue_per_method / time_by_sales_method * 1000

results_df = pd.DataFrame({
    'Total Customers': customers_per_method,
    'Total Revenue': total_revenue_per_method,
    'Average Revenue': average_revenue_per_method,
    'Time Spent (minutes)': time_by_sales_method,
    'Customer Value Ratio': CVR_per_method})
results_df.T.round(2)
```

```python
# Generate summary statistics for the 'revenue' column of a DataFrame, separated by each unique sales method.
summary_df = pd.DataFrame()
for method in df["sales_method"].unique():
    filtered_df = df[df['sales_method'] == method]["revenue"]
    summary_stats = filtered_df.describe()
    summary_df[method] = summary_stats
summary_df.round(2)
```

```python
# Create a line and scatter plot visualizing the average weekly revenue and the number of sales for each unique sales method.
plt.figure(figsize=(12, 7))
unique_methods = df['sales_method'].unique()
palette = dict(zip(unique_methods, sns.color_palette(n_colors=len(unique_methods))))
df_mean = df.groupby(['week', 'sales_method'], as_index=False).mean()
ax = sns.lineplot(data=df_mean, x='week', y='revenue', hue='sales_method', palette=palette, legend=False)
scatter = sns.scatterplot(data=df, x='week', y='revenue', hue='sales_method', size='nb_sold', sizes=(100, 400), alpha=0.8, palette=palette, ax=ax)
plt.title("Weekly Revenue Increase and Number of Sales per Sales Method", fontsize=20, pad=20)
plt.xlabel("Week")
plt.ylabel("Revenue")
handles, labels = scatter.get_legend_handles_labels()
handles = handles[1:len(unique_methods)+1]
labels = labels[1:len(unique_methods)+1]
scatter.legend(handles=handles, labels=labels, title='Sales Method', title_fontsize=14, fontsize=12, loc='upper left', bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
# Create a pivot table and a bar plot visualizing the average weekly revenue for each unique sales method.
weekly_revenue_pivot = df.pivot_table(index='week', columns='sales_method', values='revenue', aggfunc='mean')
display(weekly_revenue_pivot.round(2))

sns.set(style='whitegrid')
plt.figure(figsize=(8, 7))
g = sns.catplot(data=df, x='sales_method', y='revenue', hue='week', kind='bar', height=6, aspect=2, palette='coolwarm')
plt.title("Sales per Week per Sales Method", fontsize=20, pad=20)
plt.xlabel("Sales Method")
plt.ylabel("AVG Revenue")
plt.show()
```

```python
# Create a 2x2 grid of scatter plots with regression lines to visualize the relationships between several pairs of variables:
# Years as a customer and revenue
# Number of items sold and revenue
# Number of site visits and revenue
# Number of site visits and number of items sold

fig, axes = plt.subplots(2, 2, figsize=(14, 14))

sales_methods = df['sales_method'].unique()
colors = ['blue', 'orange', 'green']

for method, color in zip(sales_methods, colors):
    sns.regplot(ax=axes[0, 0], data=df[df['sales_method']==method], x='years_as_customer', y='revenue', color=color, label=method)
axes[0, 0].set_title('Effect of Retention on Revenue')
axes[0, 0].legend()

for method, color in zip(sales_methods, colors):
    sns.regplot(ax=axes[0, 1], data=df[df['sales_method']==method], x='nb_sold', y='revenue', color=color, label=method)
axes[0, 1].set_title('Effect of Total Sold Items on Revenue')
axes[0, 1].legend()

for method, color in zip(sales_methods, colors):
    sns.regplot(ax=axes[1, 0], data=df[df['sales_method']==method], x='nb_site_visits', y='revenue', color=color, label=method)
axes[1, 0].set_title('Effect of Site Visits on Revenue')
axes[1, 0].legend()

for method, color in zip(sales_methods, colors):
    sns.regplot(ax=axes[1, 1], data=df[df['sales_method']==method], x='nb_site_visits', y='nb_sold', color=color, label=method)
axes[1, 1].set_title('Effect of Site Visits on Sold Products')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```
